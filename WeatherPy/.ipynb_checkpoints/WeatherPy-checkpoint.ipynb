```python
#Dependencies
import json
import requests
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from citipy import citipy
from config import api_key
```

```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?units=imperial&"
column_names = ['City_ID','City', 'Cloudiness(%)','Country','Date', 'Humidity(%)','Latitude','Longitude','Max Temp(F)', 'Wind Speed(mph)']
weather_df = pd.DataFrame(columns = column_names)

# Build query URL

while len(weather_df) <500:
    lat = random.choice(range(-90, 90))
    long = random.choice(range(-180, 180))
    city = citipy.nearest_city(lat, long).city_name
    if city not in weather_df["City"].values:
        #Since the city is not in the dataframe, the url query is built
        query_url = url + "appid=" + api_key + "&q=" + city
        city_info = []
        #checks if the city is included in the weather api, if not a message is displayed
        try:
            response = requests.get(query_url).json()
            city_info.append(len(weather_df))
            city_info.append(city)
            city_info.append(response["clouds"]["all"])
            city_info.append(response["sys"]["country"])
            city_info.append(response["dt"])
            city_info.append(response["main"]["humidity"])
            city_info.append(round(response["coord"]["lat"],2))
            city_info.append(round(response["coord"]["lon"],2))
            city_info.append(response["main"]["temp_max"])
            city_info.append(response["wind"]["speed"])
            df2 = pd.DataFrame([city_info], columns=column_names)
            weather_df = weather_df.append(df2, ignore_index = True)
        except:
            print(f'{city} was not found in the Weather API')

weather_df
```

```python
#Export the city dataframe to a csv file
csv_path= "Output/cities.csv"
weather_df.to_csv(csv_path,index=False)
```

```python
#Scatter Plots
attributes = ['Max Temp(F)', 'Humidity(%)','Cloudiness(%)', 'Wind Speed(mph)']
x_values = weather_df["Latitude"]
index = 1

for attribute in attributes:
    y_values = weather_df[attribute]
    plt.scatter(x_values, y_values , marker="o", facecolors="mediumvioletred", edgecolors="c")
    plt.title(f"City Latitude vs {attribute} (Feb 2021)")
    plt.xlabel("Latitude")
    plt.ylabel(attribute)
    plt.grid()
    plt.savefig(f"Output/fig{index}.png")
    plt.show()
    index = index + 1
```

```python
#Separate the dataframe into northern and southern cities
northern_df = weather_df.loc[(weather_df["Latitude"]>=0)]
southern_df = weather_df.loc[(weather_df["Latitude"]< 0)]
```

```python
index = 1

for attribute in attributes:
    #Northern Hemisphere
    x_val1 = northern_df["Latitude"]
    y_val1 = northern_df[attribute].astype(float)
    
    #correlation
    correlation = st.pearsonr(x_val1, y_val1)
    print(f"The correlation between the northern cities & the {attribute} is {round(correlation[0],2)}")
    
    (slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val1, y_val1)
    regress_values = x_val1 * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    plt.scatter(x_val1, y_val1, marker="o", facecolors="mediumvioletred", edgecolors="c")
    plt.plot(x_val1,regress_values,"r-")
    plt.annotate(line_eq, (0,5), xycoords="figure pixels", fontsize=10,color="red")
    
    plt.title(f"Northern Hemisphere cities vs {attribute} (Feb 2021)")
    plt.xlabel("Latitude")
    plt.ylabel(attribute)
    plt.grid()
    plt.savefig(f"Output/northern_fig{index}.png")
    plt.show()
    
    #Southern Hemisphere
    
    x_val2 = southern_df["Latitude"]
    y_val2 = southern_df[attribute].astype(float)
    
    #correlation
    correlation = st.pearsonr(x_val2, y_val2)
    print(f"The correlation between the southern cities & the {attribute} is {round(correlation[0],2)}")
    
    (slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val2, y_val2)
    regress_values = x_val2 * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    plt.scatter(x_val2, y_val2, marker="o", facecolors="teal", edgecolors="powderblue")
    plt.plot(x_val2,regress_values,"r-")
    plt.annotate(line_eq, (0,5), xycoords="figure pixels", fontsize=10,color="red")
    
    plt.title(f"Southern Hemisphere cities vs {attribute} (Feb 2021)")
    plt.xlabel("Latitude")
    plt.ylabel(attribute)
    plt.grid()
    plt.savefig(f"Output/southern_fig{index}.png")
    plt.show()
    index = index + 1
```
